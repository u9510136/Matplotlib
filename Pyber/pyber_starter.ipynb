```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_path = "data/city_data.csv"
ride_path = "data/ride_data.csv"

# Read the City and Ride Data
city_csv = pd.read_csv(city_path)
ride_csv = pd.read_csv(ride_path)
# Combine the data into a single dataset
cr_csv = pd.merge(ride_csv, city_csv, how="left")
```

```python
# Display the data table for preview
cr_csv.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = cr_csv[cr_csv["type"] == "Urban"]
x_urban = urban["city"].value_counts().sort_index()
y_urban = urban.groupby("city").mean()["fare"]
size_urban = urban.groupby("city").sum()["driver_count"]

rural = cr_csv[cr_csv["type"] == "Rural"]
x_rural = rural["city"].value_counts().sort_index()
y_rural = rural.groupby("city").mean()["fare"]
size_rural = rural.groupby("city").sum()["driver_count"]

suburban = cr_csv[cr_csv["type"] == "Suburban"]
x_suburban = suburban["city"].value_counts().sort_index()
y_suburban = suburban.groupby("city").mean()["fare"]
size_suburban = suburban.groupby("city").sum()["driver_count"]

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, marker="o", facecolors="lightcoral", edgecolors="black",
            s=size_urban/3, alpha=0.75)
plt.scatter(x_rural, y_rural, marker="o", facecolors="gold", edgecolors="black",
            s=size_rural, alpha=0.75)
plt.scatter(x_suburban, y_suburban, marker="o", facecolors="dodgerblue", edgecolors="black",
            s=size_suburban/2, alpha=0.75)
# Incorporate the other graph properties
plt.grid(True)
plt.xlabel("Total Numbers of Rides(Per City)")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data(2016)")
# Create a legend
# plt.legend(("Urban","Rural","Suburban"), title="City types", loc="best")
lgnd = plt.legend(("Urban","Rural","Suburban"), title="City types", loc="best")
for handle in lgnd.legendHandles:
    handle.set_sizes([30.0])
# Incorporate a text label regarding circle size
plt.text(42,35,"Note:\nCircle size correlates with driver count per city")

# Save Figure
plt.savefig("Ride_sharing_data_2016.png",bbox_inches="tight")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_urban = urban.groupby("city").sum()["fare"].sum()
fares_rural = rural.groupby("city").sum()["fare"].sum()
fares_suburban = suburban.groupby("city").sum()["fare"].sum()
fares_percent = [fares_urban,fares_rural,fares_suburban]

# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(fares_percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)
plt.title("% of Toal Fares by City Type")

# Save Figure
plt.savefig("% of Total Fares by City Type.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_urban = urban["city"].count()
rides_rural = rural["city"].count()
rides_suburban = suburban["city"].count()
rides_percent = [rides_urban,rides_rural,rides_suburban]

# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(rides_percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=260)
plt.title("% of Toal Rides by City Type")
# Save Figure
plt.savefig("% of Total Rides by City Type.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_urban = urban.groupby("city").sum()["driver_count"].sum()
drivers_rural = rural.groupby("city").sum()["driver_count"].sum()
drivers_suburban = suburban.groupby("city").sum()["driver_count"].sum()
drivers_percent = [drivers_urban,drivers_rural,drivers_suburban]

# Build Pie Charts
labels = ["Urban", "Rural", "Suburban"]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(drivers_percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=240)
plt.title("% of Toal Drivers by City Type")

# Save Figure
plt.savefig("% of Total Drivers by City Type.png")
```

```python
# Show Figure
plt.show()
```
